### 3. Cleaning Calls Data file

```python
df=pd.read_csv(r"C:\Users\Md. Rehaan\Desktop\hackathon\calls.csv")
```

```python
df.head()
```

```python
df = df.drop('call_transcript', axis=1)
```

```python
df.head()
```

```python
# Convert the call_start_datetime to a datetime format, handling inconsistent formats
df['call_start_datetime'] = pd.to_datetime(df['call_start_datetime'], errors='coerce', dayfirst=False)

# Extract the date and time into new columns
df['call_start_date'] = df['call_start_datetime'].dt.date
df['call_start_time'] = df['call_start_datetime'].dt.time

# Display the updated dataframe with the new columns
df.head()
```

```python
# Drop rows where 'call_start_datetime' is NaT (null)
new_df = df.dropna(subset=['call_start_datetime'])

# Display the cleaned dataframe
new_df.head()
```

```python
# Convert the call_end_datetime to a datetime format, handling inconsistent formats
new_df['call_end_datetime'] = pd.to_datetime(new_df['call_end_datetime'], errors='coerce', dayfirst=False)

# Extract the date and time into new columns for call_end_datetime
new_df['call_end_date'] = new_df['call_end_datetime'].dt.date
new_df['call_end_time'] = new_df['call_end_datetime'].dt.time

# Display the updated dataframe with the new columns
new_df.head()
```

```python
cleaned_df = new_df.dropna(subset=['call_end_datetime'])

# Display the cleaned dataframe
cleaned_df.head()
```

```python
cleaned_df.to_csv("Cleaned_Calls.csv")
```
